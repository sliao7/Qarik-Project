#### Import Loan Amount Data

```python
df_cur = pd.read_pickle('currency_amounts_country.pickle')
```

```python
print("From previous steps, we have "+ str(df_cur['amount_standard'].isna().sum()) + " missing loan amount values.")
```

```python
df_cur = df_cur[['year','month','day','currency_standard','amount_standard','country_name']]
```

```python
df_cur = df_cur.dropna()
```

```python
df_cur = df_cur.reset_index(drop=True)
```

```python
df_cur.head()
```

```python
year_list = list(map(int, df_cur['year']))
year_list[0:5]
df_cur['year'] = year_list
df_cur.head()
```

```python
month_list = []
for i in range(len(df_cur)):
    if df_cur['month'].iloc[i] == 'january':
        month_list.append(int(1))
    elif df_cur['month'].iloc[i] == 'february':
        month_list.append(int(2))
    elif df_cur['month'].iloc[i] == 'march':
        month_list.append(int(3))
    elif df_cur['month'].iloc[i] == 'april':
        month_list.append(int(4))
    elif df_cur['month'].iloc[i] == 'may':
        month_list.append(int(5))
    elif df_cur['month'].iloc[i] == 'june':
        month_list.append(int(6))
    elif df_cur['month'].iloc[i] == 'july':
        month_list.append(int(7))
    elif df_cur['month'].iloc[i] == 'august':
        month_list.append(int(8))
    elif df_cur['month'].iloc[i] == 'september':
        month_list.append(int(9))
    elif df_cur['month'].iloc[i] == 'october':
        month_list.append(int(10))
    elif df_cur['month'].iloc[i] == 'november':
        month_list.append(int(11))
    elif df_cur['month'].iloc[i] == 'december':
        month_list.append(int(12))
month_list[0:5]
```

```python
df_cur['month'] = month_list
df_cur.head()
```

```python
day_list = list(map(int, df_cur['day']))
day_list[0:5]
df_cur['day'] = day_list
df_cur.head()
df_cur = df_cur.dropna()
df_cur.head()
```

```python
currency = []
for i in range(len(df_cur['currency_standard'])):
    if df_cur['currency_standard'][i] == 'us dollar':
        currency.append('USD')
    elif df_cur['currency_standard'][i] == 'euro':
        currency.append('EUR')
    elif df_cur['currency_standard'][i] == 'dem':
        currency.append('DEM')
    elif df_cur['currency_standard'][i] == 'french francs':
        currency.append('FRF')
    elif df_cur['currency_standard'][i] == 'gbp':
        currency.append('GBP')
    elif df_cur['currency_standard'][i] == 'egyptian pounds':
        currency.append('EGP')
    else:
        currency.append("")
currency[0:5]      
```

```python
len(df_cur['currency_standard'])
```

```python
len(currency)
```

```python
df_cur['currency_standard'] = currency
```

```python
df_cur = df_cur.rename(columns = {"currency_standard": "currency"})
df_cur = df_cur.rename(columns = {"amount_standard": "loan"})
df_cur = df_cur.rename(columns = {"country_name": "country"})
df_cur.head()
```

```python
df_cur['datetime'] = pd.to_datetime(df_cur[['year', 'month', 'day']]).dt.date
df_cur.head()
```

```python
df_cur = df_cur.dropna()
df_cur.head()
```

```python
df_cur['currency'].unique()
```

#### Convert to current USD

```python
date_after = datetime.date(1999, 1, 14)
```

```python
df_cur_af = df_cur[(df_cur['datetime'] > date_after)]
df_cur = df_cur.reset_index(drop=True)
df_cur_af.head()
```

```python
df_cur_af['currency'].unique()
```

```python
df_cur_af['currency'].unique()
df_cur_drop = df_cur[(df_cur['currency'] == "")]
df_cur_drop
```

```python
len(df_cur_af)
```

```python
df_new = df_cur_af.merge(df_cur_drop, how='left', indicator=True)
df_new = df_new[df_new['_merge'] == 'left_only']
```

```python
df_new = df_new[['year','month','day','currency','loan','country','datetime']]
```

```python
df_EGP = df_new[df_new['currency']=='EGP']
```

```python
df_new2 = df_new.merge(df_EGP, how='left', indicator=True)
df_new2 = df_new2[df_new2['_merge'] == 'left_only']
```

```python
df_new2.head()
```

```python
#Install package
#pip install --user currencyconverter
```

```python
from currency_converter import CurrencyConverter
c = CurrencyConverter(fallback_on_missing_rate=True)
```

```python
currency_in_usd = []
for i in range(len(df_new2)):
    date = datetime.date(df_new2['year'].iloc[i], df_new2['month'].iloc[i], df_new2['day'].iloc[i])
    dollar = c.convert(df_new2['loan'].iloc[i], df_new2['currency'].iloc[i],'USD', date)
    currency_in_usd.append(dollar)
currency_in_usd[0:5]
```

```python
df_new2['currency_in_usd'] = currency_in_usd
df_new2.head()
```

```python
df_new3 = df_new2[['year','country','currency_in_usd']]
df_new3.head()
```

```python
df_new4 = df_new3.sort_values(by='year',ascending = True).reset_index(drop=True)
df_new4.head()
```

```python
plt.figure(figsize = (16,6))
df_new3.groupby('year')['currency_in_usd'].sum().plot()
plt.ylabel('Total Amount of Loans in Nominal Values')
```

```python
cpi_list = [166.6,172.2,177.1,179.9,184.0,188.9,195.3,201.6,207.3,215.3,214.5,218.1,225,230,233,236,237,240,245,251.1,256,258.811]
```

```python
len(cpi_list)
```

```python
year_list = [i for i in range(1999,2021)]
len(year_list)
```

```python
df_CPI = pd.DataFrame({"year": year_list,"CPI": cpi_list})
```

```python
df_new5 = df_new4.merge(df_CPI, how="left", on="year")
```

```python
df_new5.head()
```

```python
df_new5['in_current_usd'] = df_new5['currency_in_usd']/df_new5['CPI']*258.811
```

```python
df_new5.head()
```

```python
plt.figure(figsize = (16,6))
df_new5.groupby('year')['in_current_usd'].sum().plot()
plt.ylabel('Total Amount of Loans in Current USD')
```
