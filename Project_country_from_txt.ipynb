## extract project and country information from the txt files

It seems that the documents start with the project description right under the headline, inside a bracket.

```python
import re
import os
import pandas as pd
```

```python
filepath = 'E:/DataScience/Qarik-Project/Qarik-Project-main/world_bank_loans_txt'

files = os.listdir(filepath)
```

```python
proj = re.compile('\((.*?)\)')
who = re.compile(', between (.*?) \(')

info = pd.DataFrame(columns=['Borrower','Project'])
bad_capture = []

for file in files:
    myfile = open(filepath+'/'+file, encoding="utf8") # open txt file for reading text
    contents = myfile.read()         # read the entire file to string
    myfile.close() 
    
    project = proj.search(contents)
    borrower = who.search(contents.replace('\n',' '))    
    
    if borrower == None and project == None:
        bad_capture.append(file)
        continue
    elif borrower == None:
        borrower_info=''
        proj_info = project.group(0).replace('(','').replace(')','')
    elif project == None:
        borrower_info=borrower.group(0).replace(', between ','').replace('(','')
        proj_info=''
    else:
        borrower_info=borrower.group(0).replace(', between ','').replace('(','')
        proj_info = project.group(0).replace('(','').replace(')','')
    
    if '(the Borrower' in borrower_info:
        borrower_info = borrower_info.replace('(the Borrower','')
    elif '("Borrower' in borrower_info:
        borrower_info = borrower_info.replace('("Borrower','')
    
    info=info.append({'Borrower':borrower_info,'Project':proj_info},ignore_index=True)
    

info.head()                          
```
